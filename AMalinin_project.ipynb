```python
def evaluate_preds(fact,prediction):
    print(f'R2: {r2(fact,prediction).round(2)}')
    
    plt.figure(figsize=(7,7))
    sns.scatterplot(x=prediction,y=fact)
    plt.plot([0,600000],[0,600000],linestyle='--',color='red')
    plt.xlabel('predicted_values')
    plt.ylabel('fact_values')
    
    plt.show()
    
    
```
